```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
import os
```

```python
PATH_UTILS        = os.path.join('..', '..', '..', 'resources', 'utils')
PATH_DESCRIPTIONS = os.path.join(PATH_UTILS, 'images_weather_description_openAI.csv')
```

```python
df = pd.read_csv(PATH_DESCRIPTIONS)
 
for index, row in df.iterrows() :
    if index > 5000 and index < 5010 :
        print(row.date_time	)
        path_img = os.path.join('..', row.path_image)
        img = mpimg.imread(path_img)
        imgplot = plt.imshow(img)
        plt.axis('off')  # Turn off axis labels
        plt.show()
        print(row.description)
        print()
```
